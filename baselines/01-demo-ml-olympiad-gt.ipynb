```python
df = pd.read_csv("../input/ml-olympiad-servicios-de-salud-guatemala/data.csv")
df.head(20)
```

```python
df_diabetes = df.loc[df.code == 'D'].copy().reset_index()
df_hipertension = df.loc[df.code == 'H'].copy().reset_index()
```

```python
X_train = df_diabetes.iloc[0:9,2:3]
y_train = df_diabetes.iloc[0:9,-1:]

X_test = df_diabetes.iloc[9:10,2:3]
y_test = df_diabetes.iloc[9:10,-1:]
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(X_train,y_train)
train_pred = reg.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error
rmse = mean_squared_error(y_train, train_pred, squared=False)
```

```python
test_pred = reg.predict(X_test)
rmse = mean_squared_error(y_test, test_pred, squared=False)
```

```python
competition_diabetes = reg.predict([[2019]])
print(competition_diabetes)
```
