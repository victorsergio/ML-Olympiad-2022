```python
# This code is the baseline code for ML-Olympiad GT 2022
# This code uses a Linear Regression using only one feature

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math

def calculate_linear_regression(df):
    
    X_train,y_train,X_val,y_val  = split_df(df)
    
    reg = LinearRegression().fit(X_train,y_train)
    
    train_pred = reg.predict(X_train)
    rmse = mean_squared_error(y_train, train_pred, squared=False)
    print("rmse: "+str(rmse))
    
    # Evaluation with 2018
    val_pred = reg.predict(X_val)
    rmse_val = mean_squared_error(y_val,val_pred, squared = False)
    print("rmse_val:" + str(rmse_val))
    
    return reg

def split_df(df):
    X_train = df.iloc[0:9,2:3]
    y_train = df.iloc[0:9,-1:]
    
    X_test = df.iloc[9:10,2:3]
    y_test = df.iloc[9:10,-1:]
    
    return X_train,y_train, X_test, y_test
    

df = pd.read_csv('../input/ml-olympiad-servicios-de-salud-guatemala/data.csv')

df_hipertension = df.loc[df.code == 'H'].copy().reset_index()
df_diabetes = df.loc[df.code == 'D'].copy().reset_index()
    
# Models    
reg_h = calculate_linear_regression(df_hipertension)
reg_d = calculate_linear_regression(df_diabetes)

# Predictions
pred_h = math.floor(reg_h.predict([[2019]]))
pred_d = math.floor(reg_d.predict([[2019]]))

# Create .csv solution file

d = {'Id': ['H','D'],'Expected':[pred_h,pred_d]}
solution_df = pd.DataFrame(data = d)
solution_df.to_csv('/kaggle/working/linear_solution.csv',index=False)

solution_df.head(5)
```
